Access NBA.com data: nba_api GitHub documentation for LeagueDashPlayerStats and LeagueDashTeamStats**

| Dataset                      | Official NBA.com page         | `nba_api` endpoint      |
| ---------------------------- | ----------------------------- | ----------------------- |
| 1. **Player Traditional Stats** | NBA Players Traditional Stats | `LeagueDashPlayerStats` |
| 2. **Player Advanced Stats**    | NBA Players Advanced Stats    | `LeagueDashPlayerStats` |
| 3. **Team Traditional Stats**   | NBA Teams Stats               | `LeagueDashTeamStats`   |
| 4. **Team Advanced Stats**      | NBA Teams Stats               | `LeagueDashTeamStats`   |




Note: NBA.com sometimes blocks or times out cloud-hosted environments. So this file is only works in google Colab but fails in Codespaces.

```python

import time
import pandas as pd
from google.colab import files

from nba_api.stats.endpoints import (
    leaguedashteamstats,
    leaguedashplayerstats
)


'''dataset player traditional'''
player_traditional = leaguedashplayerstats.LeagueDashPlayerStats(
    season="2023-24",
    season_type_all_star="Regular Season",
    per_mode_detailed="PerGame",
    measure_type_detailed_defense="Base"
).get_data_frames()[0]

player_traditional.to_csv("data/raw/nba_player_traditional_2023_24.csv", index=False)


'''dataset player advanced'''
player_advanced = leaguedashplayerstats.LeagueDashPlayerStats(
    season="2023-24",
    season_type_all_star="Regular Season",
    per_mode_detailed="PerGame",
    measure_type_detailed_defense="Advanced"
).get_data_frames()[0]

player_advanced.to_csv("data/raw/nba_player_advanced_2023_24.csv", index=False)


'''dataset team traditional'''
team_traditional = leaguedashteamstats.LeagueDashTeamStats(
    season="2023-24",
    season_type_all_star="Regular Season",
    per_mode_detailed="PerGame",
    measure_type_detailed_defense="Base"
).get_data_frames()[0]

team_traditional.to_csv("data/raw/nba_team_traditional_2023_24.csv", index=False)


'''dataset team advanced'''
team_advanced = leaguedashteamstats.LeagueDashTeamStats(
    season="2023-24",
    season_type_all_star="Regular Season",
    per_mode_detailed="PerGame",
    measure_type_detailed_defense="Advanced"
).get_data_frames()[0]

team_advanced.to_csv("data/raw/nba_team_advanced_2023_24.csv", index=False)


############ download to local ###############
files.download("data/raw/nba_player_traditional_2023_24.csv")
files.download("data/raw/nba_player_advanced_2023_24.csv")
files.download("data/raw/nba_team_traditional_2023_24.csv")
files.download("data/raw/nba_team_advanced_2023_24.csv")

```
